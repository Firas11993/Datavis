```python
import pandas as pd 
import numpy as np
from datetime import datetime
```

```python
DATA_DIR = "data"
```

```python
stops = pd.read_csv(DATA_DIR + "/ter/stops.txt")
stop_times = pd.read_csv(DATA_DIR + "/ter/stop_times.txt")
```

```python
merged = pd.merge(stop_times,stops,how='left', left_on=stop_times.stop_id,right_on = stops.stop_id) \
            [['trip_id','arrival_time','departure_time','stop_name','stop_lat','stop_lon','stop_sequence']]
```

```python
grouped = merged.copy()
```

```python
def fix_times(time) : 
    
    if int(time.split(":")[0]) >= 24:
        return "{}:".format(int(time.split(":")[0]) - 24)+":".join(time.split(":")[1:])
    return time
merged['departure_time'] = merged['departure_time'].apply(fix_times)
merged['arrival_time'] = merged['arrival_time'].apply(fix_times)
```

```python
timetable = merged.groupby("trip_id").agg(list).reset_index()
timetable
```

```python
duration=[]
for i in range (len(timetable)):
    duration.append(pd.to_datetime(timetable['arrival_time'][i][-1]) - pd.to_datetime(timetable['departure_time'][i][0]))
```
